```python
from collections import Counter

l = [tuple(v) for v in df_multil.to_list()]

k = Counter(l).keys()
v = Counter(l).values()

fr = {x:y for y, x in sorted(zip(v, k), reverse=True)}

labels = df.columns[2:12]
cat_dict = dict()

for x, y in fr.items():
    new_k = []
    for val in x:
        new_k.append(labels[val])
    cat_dict[tuple(new_k)] = y

cat_dict
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained("Hate-speech-CNERG/dehatebert-mono-german")
```

```python
d = df.dropna()
subset = d["kommentar"].tolist()[0:100]
labels = d['label'].tolist()[0:100]

s_t, s_d = [], []
for v in subset:
    e = tokenizer.encode(v, padding=True)
    d = tokenizer.decode(e)
    s_t.append(e)
    s_d.append(d)
```

```python
outarr=np.ones((np.max([len(ps) for ps in s_t]),len(s_t)))*0 #define empty array
for i,c in enumerate(s_t):  #populate columns
    outarr[:len(c),i]=c
s_t = outarr.transpose()
labels = np.array(labels)
print(s_t.shape)
print(labels.shape)
```

```python
from imblearn.over_sampling import SMOTE

X, y = s_t, labels
print('Original dataset shape %s' % Counter(y))
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))
```

```python
for val in X_res:
    i_val = [int(v) for v in val]
    print(tokenizer.decode(i_val))
```
